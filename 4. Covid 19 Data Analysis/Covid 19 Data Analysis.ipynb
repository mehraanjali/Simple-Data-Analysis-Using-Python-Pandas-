#### Q1. Show the number of confirmed, Deaths and recovered cases in each region.

```python
df.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head()
```

```python
df.groupby('Region')['Deaths'].sum().sort_values(ascending = False).head(10)
```

```python
df.groupby('Region')['Recovered'].sum().sort_values(ascending = False).head(10)
```

### OR

```python
df.groupby('Region')['Confirmed','Deaths','Recovered'].sum().head(10)
```

#### Q2. Remove all the records where confirmed cases is less than 10.

```python
df = df[~(df.Confirmed < 10)]
df
```

#### Q3. In which Region, maximum number of confirmed cases were recorded?

```python
df.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(1)
```

#### Q4. In which region, minimum number of deaths cases were recorded?

```python
df.groupby('Region')['Deaths'].sum().sort_values(ascending = True).head(30)
```

#### Q5. How many confirmed, deaths & recovered cases were reported from India till 29 april 2020?

```python
df[df.Region == 'India']
```

#### Q6. A) Sort the entire data wrt No. of confirmed cases in asc order.

```python
df.sort_values(by= ['Confirmed'], ascending = True)
```

#### Q6. B) Sort the entire data wrt No. of recovered cases in desc order.
