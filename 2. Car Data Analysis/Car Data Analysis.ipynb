# Cars Dataset

```python
import pandas as pd
df = pd.read_csv('Car Dataset.csv')
df
```

### 1) Data Cleaning

##### Find the null values if any, and fill it with the mean of that column.

```python
df.isnull().sum()
```

```python
# Remove blank rows (rows with all column values as blank)
df.dropna(how='all', inplace=True)
```

```python
df.isnull().sum()
```

```python
# If you want to fill missing values in all columns with their respective means, you can use:
#df_cleaned = df.apply(lambda col: col.fillna(col.mean()), axis=0)
```

```python
# Fill missing values in a specific column (replace 'your_column' with the actual column name)
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

### 2) Value Based on counts

##### Check what are the different types of make are there in our dataset. And, what is the count (occurrence) of each Make in the data?

```python
df['Make'].value_counts()
```

### 3) Instruction (Filtering)

##### Show all the records where Origin is Asia or Europe.

```python
df[df['Origin'].isin(['Asia','Europe'])]
```

### 4) Instruction (Removing unwanted records) 

##### Remove all the records (rows) where weight is above 4000.

```python
# Remove rows where the weight is above 4000
#df = df[df['Weight'] <= 4000]
#df

# Remove rows where the weight is above 4000
df = df.drop(df[df['Weight'] > 4000].index)
df
```

### 5) Instruction (Applying function on a column)

##### Increase all the values of 'MPG_City' column by 3.

```python
df['MPG_City']
```

```python
df['MPG_City'] = df['MPG_City'].apply(lambda x:x+3)
```
